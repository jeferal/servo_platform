```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from copy import deepcopy

from time import sleep

import rospy
import rosbag

from cv_bridge import CvBridge
```

```python
# Load the rosbag and create an array of images from the topic /camera/rgb/image_raw
bag = rosbag.Bag('/home/user/projects/shadow_robot/base/src/servo_platform/sp_perception/data/2023-09-02-10-23-36.bag')
images = []

bridge = CvBridge()

for topic, msg, t in bag.read_messages(topics=['/usb_cam/image_raw']):
    cv_image = bridge.imgmsg_to_cv2(msg, desired_encoding='mono8')
    images.append(cv_image)
```

```python
def show_video(images, process_image=None):
    for image in images:
        if process_image is not None:
            image = process_image(image)
        cv2.imshow('image', image)
        cv2.waitKey(1)
        sleep(0.03333333333)
    cv2.destroyAllWindows()
```

```python
show_video(images)
```

```python
# Apply a gaussian filter to the images
gaussian_images = []
for image in images:
    gaussian_images.append(cv2.GaussianBlur(image, (5, 5), 0))
```

```python
show_video(gaussian_images)
```

```python
def process_image(image):
    # Create a copy of the image
    # original_image = deepcopy(image)

    # Apply gaussian filter
    image = cv2.GaussianBlur(image, (5, 5), 10)

    # Increase the contrast
    image = cv2.equalizeHist(image)

    # Detect circles
    # circles = cv2.HoughCircles(image, cv2.HOUGH_GRADIENT, dp=1, minDist=50, param1=127, param2=25, minRadius=25, maxRadius=80)
    circles = cv2.HoughCircles(image, cv2.HOUGH_GRADIENT, dp=1, minDist=50, param1=100, param2=30, minRadius=25, maxRadius=70)

    # Print number of circles
    if circles is not None:
        print('Number of circles detected: {}'.format(circles.shape[1]))
    else:
        print('No circles detected')

    # Draw circles
    if circles is not None:
        circles = np.uint16(np.around(circles))
        circle = circles[0][0]
        cv2.circle(image, (circle[0], circle[1]), circle[2], (0, 255, 0), 2)
        cv2.circle(image, (circle[0], circle[1]), 2, (0, 0, 255), 3)

    return image
```
